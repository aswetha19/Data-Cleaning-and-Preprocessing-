```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\pslas\OneDrive\Desktop\Day-1.csv")
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['Gender'] = df['Gender'].str.lower().str.strip()
df['Gender'].replace({'f':'female','m':'male'}, inplace=True)
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_', regex=False)
```

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'], errors='coerce')
df['appointmentday'] = pd.to_datetime(df['appointmentday'], errors='coerce')
df
```

```python
df['patientid'] = df['patientid'].astype('int64').astype(str)
df
```

```python
df['age'] = df['age'].astype('int64')
```

```python
df.to_csv(r"C:\Users\pslas\OneDrive\Desktop\Day-1.csv",index=False)
df
```

```python
df = df[(df['age'] >= 0) & (df['age'] <= 120)]
df
```

```python
df['neighbourhood'] = df['neighbourhood'].str.strip().str.title()
df
```

```python
df['scheduled_day'] = df['scheduledday'].dt.day_name()
df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days
df
```
